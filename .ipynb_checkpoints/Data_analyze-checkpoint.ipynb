```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

df = pd.read_csv('AppleStore.csv')

print(df.head())
```

```python
print(df.describe())
```


let's count the prime genre


```python

genre = df['prime_genre'].tolist()

wordcount = dict((x,0) for x in genre)
for w in genre:
    if w in wordcount:
        wordcount[w] += 1
dict_genre = wordcount
print(dict_genre)
#print(type(wordcount))


df_genre2 = pd.DataFrame(list(wordcount.iteritems()),columns=['Genre','Count'])
df_genre2 = df_genre2.sort_values(by=['Count'])


# plot
df_genre2.plot.barh(x='Genre',y='Count',figsize=(10,8), legend=False)
# add a suptitle
plt.suptitle("Genre of app in applestore", fontsize=18)
# set xlabel to ""
plt.xlabel("")
# change xticks fontsize to 14
plt.yticks(fontsize=14)
# finally show the plot
plt.show()
```


Lot of game , it is not a big surprise indeed. let's check the currency to have an indication of where it is bought.


```python

currency = df['currency'].tolist()

wordcount = dict((x,0) for x in currency)
for w in currency:
    if w in wordcount:
        wordcount[w] += 1
        
print(wordcount)

```

Good news we can use the price without doing any kind of conversion.

```python
# let's check the price to have an indication of where it is bought
price = df['price'].tolist()

wordcount = dict((x,0) for x in price)
for w in price:
    if w in wordcount:
        wordcount[w] += 1

df_genre3 = pd.DataFrame(list(wordcount.iteritems()),columns=['Price','Count'])
df_genre3 = df_genre3.sort_values(by=['Price'])

print(df_genre3)

#sns.boxplot(x,orient=None, color=None, saturation=2, width=2, fliersize=5, linewidth=None, whis=1.5)
```

there is not a lot of app price above 20, lets zoom under 20

```python

filter = df['price']<20
df1 = df[filter]
g = sns.jointplot(x=df1['user_rating'], y=df1['price'], data=df1, kind="kde", color="m",figsize = (12,8))
g.plot_joint(plt.scatter, c="b", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("user rating", "price");
```

```python

filter = df['price'] == 0
df2 = df[filter]

df_price_free = pd.DataFrame(df2)
df_price_free = df_price_free.sort_values(by=['rating_count_tot'],ascending=False)

filter = df['price'] != 0
df3 = df[filter]

df_price_notfree = pd.DataFrame(df3)
df_price_notfree = df_price_notfree.sort_values(by=['rating_count_tot'],ascending=False)
print('FREE APP :')
print(df_price_free.head())
print('-------------------------------------------------------------------------------------------------------------')
print('PAYING APP :')
print(df_price_notfree.head())
```

Here we go, time to prediction we will aim for a good user_rating which is a criteria of success, we will look on rating_count_tot after.
Let's drop some predictor as rating_count_ver, user_rating_ver or rating_count_tot which are a consequence of the success of an application


```python

df = pd.read_csv('AppleStore.csv')

rating = df['user_rating'].tolist()

wordcount = dict((x,0) for x in rating)
for w in rating:
    if w in wordcount:
        wordcount[w] += 1
dict_rating = wordcount



list_rating = list()
for i in dict_rating.keys():
    list_rating.append(i)

df_rating = pd.DataFrame(list_rating,columns=['rating'])
df_rating['size'] = float(0)



for i in range(len(df)):
        for a in range(len(df_rating)):
            if (df.iloc[i,8]  == df_rating.iloc[a,0]):
                df_rating.iloc[a,1] = df_rating.iloc[a,1] + df.iloc[i,3]
                
                
for keys,values in dict_rating.items():
    for i in range(len(df_rating)):
        if(keys == df_rating.iloc[i,0]):
            df_rating.iloc[i,1] = df_rating.iloc[i,1]/values
            
df_rating.sort_values(by='rating')
```

```python
from math import *
print(type(df_rating['size']))


f, ax = plt.subplots(1,1,figsize = (12,8))
g = sns.barplot(x = df_rating['rating'], y = df_rating['size'], orient = 'w', palette="Blues_d")
#g.set(yscale="log")
ax.set_title('Average rating by size', fontsize = 14)
plt.show()
```

Looks like quality and quantity are meaningful till a certain point. arount 200MB is a good size.But we can be more accurate.

```python
from wordcloud import WordCloud

# let's check on the description of these famous applications?
df_description = pd.read_csv('appleStore_description.csv')
df = pd.read_csv('AppleStore.csv')

df_description = df_description.dropna(axis=1)
df = df.merge(df_description, on='id')



# Visualize data with word Cloud
def visualize(rank):
    words = ''
    for msg in df['app_desc'].head(rank):
        msg = msg.lower()
        words += msg+' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title('Top 500 popular applications')
    plt.show()

visualize(500)
```
