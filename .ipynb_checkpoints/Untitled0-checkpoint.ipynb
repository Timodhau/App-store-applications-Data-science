```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

df = pd.read_csv('AppleStore.csv')

print(df.head())
```


let's count the prime genre


```python

genre = df['prime_genre'].tolist()

wordcount = dict((x,0) for x in genre)
for w in genre:
    if w in wordcount:
        wordcount[w] += 1
dict_genre = wordcount
print(dict_genre)
#print(type(wordcount))


df_genre2 = pd.DataFrame(list(wordcount.iteritems()),columns=['Genre','Count'])
df_genre2 = df_genre2.sort_values(by=['Count'])


# plot
df_genre2.plot.barh(x='Genre',y='Count',figsize=(10,8), legend=False)
# add a suptitle
plt.suptitle("Genre of app in applestore", fontsize=18)
# set xlabel to ""
plt.xlabel("")
# change xticks fontsize to 14
plt.yticks(fontsize=14)
# finally show the plot
plt.show()
```

Now let's see if we can get something with predictions.

```python
# 

from sklearn.preprocessing import LabelEncoder
var_mod = ['size_bytes','currency','price','rating_count_tot','rating_count_ver','user_rating','user_rating_ver','ver', 'cont_rating','prime_genre','sup_devices.num','lang.num']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
#Here we go, time to prediction we will aim for a good user_rating which is a criteria of success, we will look on rating_count_tot after.
#Let's drop some predictor as rating_count_ver, user_rating_ver or rating_count_tot which are a consequence of the succes of an application


#Import models from scikit learn module:
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics

#Generic function for making a classification model and accessing performance:
def classification_model(model, data, predictors, outcome):
  #Fit the model:
  model.fit(data[predictors],data[outcome])
    
#We can try different combination of variables:
model = RandomForestClassifier(n_estimators=50)
outcome_var = 'user_rating'
predictor_var = ['size_bytes','price','ver', 'cont_rating','prime_genre','sup_devices.num','lang.num']
classification_model(model, df,predictor_var,outcome_var)

#Let's drop some predictor as rating_count_ver, user_rating_ver or rating_count_tot which are a consequence of the succes of an application


#Make predictions on training set:
predictions = model.predict(df[predictor_var])

#Print accuracy
accuracy = metrics.accuracy_score(predictions,df[outcome_var])
print "Accuracy : %s" % "{0:.3%}".format(accuracy)


#Perform k-fold cross-validation with 5 folds
kf = KFold(df.shape[0], n_folds=5)
error = []
for train, test in kf:
    # Filter training data
    train_predictors = (df[predictor_var].iloc[train,:])
    
    # The target we're using to train the algorithm.
    train_target = df[outcome_var].iloc[train]
    
    # Training the algorithm using the predictors and target.
    model.fit(train_predictors, train_target)
    
    #Record error from each cross-validation run
    error.append(model.score(df[predictor_var].iloc[test,:], df[outcome_var].iloc[test]))
 
print "Cross-Validation Score : %s" % "{0:.3%}".format(np.mean(error))

#Fit the model again so that it can be refered outside the function:
model.fit(df[predictor_var],df[outcome_var]) 

#Create a series with feature importances:
featimp = pd.Series(model.feature_importances_, index=predictor_var).sort_values(ascending=False)
print featimp

featimp.plot(kind='bar', title='Feature Importances')
```

The cross validation is way too low to use this model. But it gives us a good idea. Look the effect of the size on the rating.

```python

df = pd.read_csv('AppleStore.csv')

rating = df['user_rating'].tolist()

wordcount = dict((x,0) for x in rating)
for w in rating:
    if w in wordcount:
        wordcount[w] += 1
dict_rating = wordcount



list_rating = list()
for i in dict_rating.keys():
    list_rating.append(i)

df_rating = pd.DataFrame(list_rating,columns=['rating'])
df_rating['size'] = float(0)



for i in range(len(df)):
        for a in range(len(df_rating)):
            if (df.iloc[i,8]  == df_rating.iloc[a,0]):
                df_rating.iloc[a,1] = df_rating.iloc[a,1] + df.iloc[i,3]
                
                
for keys,values in dict_rating.items():
    for i in range(len(df_rating)):
        if(keys == df_rating.iloc[i,0]):
            df_rating.iloc[i,1] = df_rating.iloc[i,1]/values
            
df_rating.sort_values(by='rating')
```

```python
from wordcloud import WordCloud

# let's check on the description of these famous applications?
df_description = pd.read_csv('appleStore_description.csv')
df = pd.read_csv('AppleStore.csv')

df_description = df_description.dropna(axis=1)
df = df.merge(df_description, on='id')



# Visualize data with word Cloud
def visualize(rank):
    words = ''
    for msg in df['app_desc'].head(rank):
        msg = msg.lower()
        words += msg+' '
    wordcloud = WordCloud(width=600, height=400).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title('Top 500 popular applications')
    plt.show()

visualize(500)
```

It is interesting to see some words disappearing between the description. Free is one of them.
